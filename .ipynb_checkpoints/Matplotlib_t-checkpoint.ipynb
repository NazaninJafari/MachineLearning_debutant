```python
x1= np.array([2,3,4,5])
y1= np.array([10,2,4,6])
plt.subplot(2,3,1)
plt.plot(x1,y1,'--g')
plt.title('samdi')

x2= np.array([1,2,3,4])
y2= np.array([2,6,3,5])
plt.subplot(2,3,2)
plt.plot(x2,y2,'--y')
plt.title('dimanche')

plt.subplot(2,3,3)
plt.plot(x1,y1,'--g')
plt.title('lundi')

plt.subplot(2,3,4)
plt.plot(x2,y2,'--y')
plt.title('mardi')

plt.subplot(2,3,5)
plt.plot(x1,y1,'--g')
plt.title('mercredi')

plt.subplot(2,3,6)
plt.plot(x2,y2,'--y')
plt.title('jeudi')

plt.suptitle('taux de vente')
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c= colors)
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x,y, c=colors, cmap='viridis')

plt.colorbar()
plt.show()
```

```python
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)
plt.show()
```

```python
x = np.random.randint(100, size=100)
y = np.random.randint(100, size=100)
colors = np.random.randint(100, size=100)
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, cmap='nipy_spectral', s=sizes, alpha=0.5 )
plt.colorbar()
plt.show()
```

```python
x= np.array([10,45,60,30])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(x, labels=mylabels)
plt.show()
```

```python
plt.pie(x, labels=mylabels, startangle=90)
plt.show()
```

```python
mycolors = ["black", "hotpink", "b", "#4CAF50"]
myexplode = np.array([0,0.1,0,0])
plt.pie(x, labels=mylabels, colors= mycolors, explode = myexplode, shadow = True)

plt.show()
```
